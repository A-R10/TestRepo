Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Example: Find the distance from a launch site to a city

# Choose a launch site (e.g., KSC LC-39A)
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# Coordinate of a city point (replace with your observed coordinates from MousePosition)
city_lat = 28.6114
city_lon = -80.8208 # Example coordinates for Titusville, FL

# Calculate the distance
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

print(f"Distance from KSC LC-39A to city: {distance_city:.2f} KM")

# Create a marker for the city point with the distance
distance_marker_city = folium.Marker(
   [city_lat, city_lon],
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html=f'<div style="font-size: 12; color:#d35400;"><b>{distance_city:.2f} KM</b></div>',
       )
   )
site_map.add_child(distance_marker_city)

# Create a `folium.PolyLine` object between the launch site and city
city_coordinates = [[launch_site_lat, launch_site_lon], [city_lat, city_lon]]
lines_city=folium.PolyLine(locations=city_coordinates, weight=1)
site_map.add_child(lines_city)

# Display the map
site_map
```
